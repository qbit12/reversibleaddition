# Simulating Reversible Addition

```python
from qiskit import *
```

```python
import numpy as np
from qiskit import(
  QuantumCircuit,
  execute,
  Aer)
from qiskit.visualization import plot_histogram
```

```python

def add(circuit,a,b,c,d):
  circuit.ccx(b,c,d)
  circuit.cx(b,c)
  circuit.ccx(a,c,d)
```

```python
def plus(circuit,a,b,c):
  circuit.cx(b,c)
  circuit.cx(a,c)
```

```python
simulator = Aer.get_backend('qasm_simulator')
circuit = QuantumCircuit(12,8)


circuit.h(1)
circuit.h(2)

circuit.h(3)
circuit.h(4)
circuit.h(6)
circuit.h(7)
circuit.h(0)
circuit.h(9)



add(circuit,1,2,3,4,)

add(circuit,3,4,5,6)

add(circuit,6,7,8,9)
circuit.cx(7,8)
plus(circuit,6,7,8)

add(circuit,3,4,5,6)
plus(circuit,3,4,5)
add(circuit,0,1,2,3)
plus(circuit,0,1,2)

circuit.measure([2,5,8,9], [0,1,2,3])

# Execute the circuit on the qasm simulator
job = execute(circuit, simulator, shots=100000)
result = job.result()
counts = result.get_counts(circuit)
```

```python
%matplotlib inline
circuit.draw(output = "mpl")
```

```python
plot_histogram(counts)
```
